```python
import os
import gymnasium as gym
import numpy as np
from stable_baselines3 import SAC
import mediapy as media

import rl_envs
import warnings
import logging

# Silence warnings
warnings.filterwarnings("ignore")
os.environ.setdefault("PYTHONWARNINGS", "ignore")

# Reduce noisy log output from common libraries
for logger_name in ("gymnasium", "stable_baselines3", "rl_envs", "matplotlib", "urllib3"):
    logging.getLogger(logger_name).setLevel(logging.ERROR)

# ---- Paths ----
algo_env = "SAC_RampPush-v0"
run_id = f"20251124-173205" 
base_log = f".logs/{algo_env}/{run_id}"

best_model_path = os.path.join(base_log, "models", "best_model")
video_path = os.path.join(base_log, "videos", "eval_rollout.mp4")

print("Loading:", best_model_path)
model = SAC.load(best_model_path)

# ---- Create env with rgb_array mode ----
env = gym.make("RampPush-v0", render_mode="rgb_array")

frames = []
obs, info = env.reset()

done = False
truncated = False

while not (done or truncated):
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, truncated, info = env.step(action)
    frame = env.render()            # returns H x W x 3 numpy array
    frames.append(frame)

env.close()

media.show_video(frames)
```
